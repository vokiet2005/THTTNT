````python
class Graph:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis

    def get_neighbors(self, v):
        return self.adjac_lis.get(v, [])

    def h(self, n):
        # This is a placeholder heuristic function.
        # For a general A* algorithm, this should return an estimated cost from node 'n' to the goal.
        # A common simple heuristic is 0, which makes A* behave like Dijkstra's algorithm.
        return 0

    def a_star_algorithm_N_nodes(self, start, N):
        open_lst = set([start])
        closed_lst = set([])

        # g(n): chi ph√≠ t·ª´ start t·ªõi n
        poo = {} # 'poo' likely means 'g' cost or cost_so_far
        poo[start] = 0

        # l∆∞u cha ƒë·ªÉ truy v·∫øt ƒë∆∞·ªùng ƒëi
        par = {} # 'par' likely means 'parent'
        par[start] = start

        while len(open_lst) > 0:
            n = None

            # ch·ªçn ƒë·ªânh c√≥ f(n) = g(n) + h(n) nh·ªè nh·∫•t
            for v in open_lst:
                if n is None or poo[v] + self.h(v) < poo[n] + self.h(n):
                    n = v

            if n is None:
                return None

            # ===== KI·ªÇM TRA ƒê·ª¶ N ƒê·ªàNH =====
            count = 1
            temp = n
            while par[temp] != temp:
                count += 1
                temp = par[temp]

            if count == N:
                path = []
                while par[n] != n:
                    path.append(n)
                    n = par[n]
                path.append(start)
                path.reverse()
                return path

            # duy·ªát c√°c ƒë·ªânh k·ªÅ
            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    poo[m] = poo[n] + weight
                else:
                    if poo[m] > poo[n] + weight:
                        poo[m] = poo[n] + weight
                        par[m] = n
                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        return None

adjac_lis = {
    'A': [('B', 1), ('C', 1)],
    'B': [('D', 1), ('E', 1)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}
g = Graph(adjac_lis)

path = g.a_star_algorithm_N_nodes('B', 2)
print(path)
````
